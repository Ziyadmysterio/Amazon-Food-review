```python
#Module to perform text pre processing and Normalizations


# IMPORTS
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords

import spacy
from nltk.stem.porter import PorterStemmer

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

stop = stopwords.words('english')



###  FUNCTIONS   ###

# Function to convert all upper case text into lower case
def To_Lower(df_reviews):
    df_reviews['Summary_text']=df_reviews['Summary_text'].str.lower()
    return df_reviews

# Function to remove all HTML break tags 
def Remove_HTML_Tag(df_reviews):
    df_reviews['Summary_text']=[doc.replace("<br />", " ") for doc in df_reviews.Summary_text]
    return df_reviews


# Function to remove all Special Characters from text
def Remove_Special_Char(df_reviews):
    df_reviews['Summary_text']=df_reviews['Summary_text'].str.replace('[^\w\s]','')
    return df_reviews

# Function for removing the stopword
def stopwords(Summary_text):
    
    global stop
    # removing the stop words and lowercasing the selected words
    text = [word.lower() for word in Summary_text.split() if word.lower() not in stop]
    # joining the list of words with space separator
    return " ".join(text)

# Function to remove stopwords
def Remove_Stopwords(df_reviews):
    df_reviews['Summary_text'] = df_reviews['Summary_text'].apply(stopwords)
    return df_reviews



# A function to perform all preprocessing tasks 
def Perform_Preprocessing(df_reviews):
    df_reviews=To_Lower(df_reviews)
    df_reviews=Remove_HTML_Tag(df_reviews)
    df_reviews=Remove_Special_Char(df_reviews)
    df_reviews=Remove_Stopwords(df_reviews)
    return df_reviews

# A function to perform stemming Normalization
def Perform_Stemming(reviews):
    
    stemmer = PorterStemmer()
    return [' '.join([stemmer.stem(word) for word in review.split()]) for review in reviews]
    

    
def lemmatize_text(review):
    w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
    lemmatizer = WordNetLemmatizer()
    text=[ lemmatizer.lemmatize(word) for word in w_tokenizer.tokenize(review) ]
    return " ".join(text)

# A function to perform Lemmatization Normalization
def Perform_Lemmatization(df_reviews):
    df_reviews['Summary_text'] = df_reviews['Summary_text'].apply(lemmatize_text)
    return df_reviews

    
    
    
   



```
